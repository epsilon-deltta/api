numpy loadtxt는 줄별로 읽으며 필요한 수치로 나누어 불러오는 것이 가능합니다. 다만 이는 array 형식이며 테이블 데이터에 최적화되어있지는 않습니다.  
numpy는 애초에 선형수학의 도구로 쓰이는 편이니, 그에 맞게 조정하는데 시간이 쓸릴 가능성이 큽니다

```python
import pandas as pd
col_name = ['transName','transID','chrom','strand','txnSta','txnEnd','cdsSta','cdsEnd','exnNum','exnSta','exnEnd']
refFlat_DF = pd.read_csv('./refFlat.txt', sep='\t', names=col_name, header=None)
refFlat_DF
```

이 경우 다른건 다 좋으나 일부 문자열인 부분이 마음에 걸립니다

```python
refFlat_DF[['exnSta', 'exnEnd']]
```

쉼표를 기준으로 숫자를 구분해야 하므로 맵핑 함수를 씁니다

```python
refFlat_DF['exnSta'] = refFlat_DF['exnSta'].apply(np.fromstring, sep=',', dtype=int)
refFlat_DF['exnEnd'] = refFlat_DF['exnEnd'].apply(np.fromstring, sep=',', dtype=int)
refFlat_DF[['exnSta','exnEnd']]
```

만일 갯수를 세고 싶다면 apply를 다시 이용하면 됩니다

```python
refFlat_DF['exnSta'].apply(len)
```

exnSta와 exnEnd를 하나씩 짝짓고 싶으면 다음과 같은 방식으로 하면 됩니다  
exnSta와 exnEnd에서 순차적으로 가져온 리스트가 각각 x,y로 지정되고  
x,y를 zip[리스트 내 쌍을 튜플로 묶어 하나씩 반환]으로 한 뒤 리스트로 맞추고 이를 적용한다는 map을 이용  
마지막으로 row 당 하나의 리스트로 만들려고 마지막에 list를 적용하면 됩니다

```python
print(refFlat_DF['exnSta'].loc[0], refFlat_DF['exnEnd'].loc[0])
list(zip(refFlat_DF['exnSta'].loc[0], refFlat_DF['exnEnd'].loc[0]))
```

```python
list(map(lambda x,y : list(zip(x,y)), refFlat_DF['exnSta'], refFlat_DF['exnEnd']))
```

```python
refFlat_DF['exnList'] = list(map(lambda x,y : list(zip(x,y)), refFlat_DF['exnSta'], refFlat_DF['exnEnd']))
refFlat_DF['exnList']
```

numpy계산이 더 편하다 하면 이와 같은 방법도 있습니다

```python
refFlat_DF['exnList'] = list(map(lambda x,y : np.transpose([x,y]), refFlat_DF['exnSta'], refFlat_DF['exnEnd']))
refFlat_DF['exnList']
```

exnList에서 각각의 길이를 알고 싶으면 차이값을 구하면 되므로 np.diff함수를 쓰면 됩니다  
또한 길이의 합은 sum함수로 구하면 편합니다

```python
refFlat_DF['exnLenList'] = refFlat_DF['exnList'].apply(np.diff)
refFlat_DF['exnLen'] = refFlat_DF['exnLenList'].apply(sum)
refFlat_DF[['exnLenList','exnLen']]
```

apply에 커스텀 프로세스로 복잡한 계산을 하고 싶다면 우선 함수를 정의해야합니다  
이에 예시로는 cdsSta 및 cdsEnd를 기준으로 exnList를 나누는 것을 해보겠습니다  
편의 상 numpy array로 합니다

```python
# 입력 값은 refFlat_DF의 strand, exnSta, exnEnd, exnList, cdsSta, cdsEnd
def sep_exn(F_strand, F_exnSta, F_exnEnd, F_exnList, F_cdsSta, F_cdsEnd):
    # intron 구하기
    F_intron = np.transpose([F_exnEnd[:-1], F_exnSta[1:]])
    # cdsSta와 cdsEnd가 같으면 패스
    if F_cdsSta != F_cdsEnd:
        # exn에서 cds의 위치 파악
        front_index = np.searchsorted(F_exnEnd,F_cdsSta)+1
        back__index = np.searchsorted(F_exnSta,F_cdsEnd)+1
        # cds를 두개 씩 아래에서 쑤셔넣고 정렬하여 나누기 쉽게
        F_exnList = np.sort(np.concatenate((F_exnList, [[F_cdsSta,F_cdsSta],[F_cdsEnd,F_cdsEnd]])),axis=0)
        # cds 중간 추출
        F_middl_lst = F_exnList[front_index:back__index]
        # strand에 따라 앞뒤 추출
        if F_strand == '+':
            F_front_lst = F_exnList[:front_index]
            F_back__lst = F_exnList[back__index:]
        else:
            F_back__lst = F_exnList[:front_index]
            F_front_lst = F_exnList[back__index:]
        # 나눈 리스트에서 잘라지지 않을 경우에 대해 보정 및 각각의 길이랑 전체길이
        def LLS_cal(FF_lst):
            FF_len = np.diff(FF_lst)
            FF_TF = FF_len.astype(bool).flatten()
            FF_lst = FF_lst[FF_TF]
            FF_len = FF_len[FF_TF]
            FF_sum = sum(FF_len)
            return FF_len, FF_lst, FF_sum
        #함수 적용
        F_front_len, F_front_lst, F_front_sum = LLS_cal(F_front_lst)
        F_middl_len, F_middl_lst, F_middl_sum = LLS_cal(F_middl_lst)
        F_back__len, F_back__lst, F_back__sum = LLS_cal(F_back__lst)
    # 같을 경우는 중간만 입력받고 내뱉기
    else:
        F_front_lst, F_back__lst = [],[]
        F_middl_lst = F_exnList
        F_middl_len = np.diff(F_middl_lst)
        F_middl_sum = sum(F_middl_len)
        F_front_len, F_back__len = [],[]
        F_front_sum, F_back__sum = 0, 0
    # 처리 결과를 리스트로 반환
    return [F_front_lst, F_front_len, F_front_sum, F_middl_lst, F_middl_len, F_middl_sum, F_back__lst, F_back__len, F_back__sum, F_intron]
```

이걸 토대로 해서 나눈 것이 다음 방식입니다

```python
F_middl_lst = F_exnList[front_index:back__index]
if F_strand == '+':
    F_front_lst = F_exnList[:front_index]
    F_back__lst = F_exnList[back__index:]
else:
    F_back__lst = F_exnList[:front_index]
    F_front_lst = F_exnList[back__index:]
```

다만 만약 [(1,2)(4,5)(7,8)]일 경우 분할이 3, 6일 시  
[(1,2)(3,3)(4,5)(6,6)(7,8)]이 될 수 있으므로 이를 해결하기 위해 len으로 길이를 구하고 그것이 0일 경우 boolen으로 제거하는 방식을 포함한 것이

```python
def LLS_cal(FF_lst):
    FF_len = np.diff(FF_lst)              # 길이 구하기
    FF_TF = FF_len.astype(bool).flatten() # 길이가 0이면 False
    FF_lst = FF_lst[FF_TF]                # 길이가 False인 경우 제거
    FF_len = FF_len[FF_TF]                # 길이가 False인 경우 제거
    FF_sum = sum(FF_len)                  # 길이 합
    return FF_len, FF_lst, FF_sum
```

위 함수입니다

```python
F_front_len, F_front_lst, F_front_sum = LLS_cal(F_front_lst)
F_middl_len, F_middl_lst, F_middl_sum = LLS_cal(F_middl_lst)
F_back__len, F_back__lst, F_back__sum = LLS_cal(F_back__lst)
```

```python
ex_ref_DF = [F_front_lst, F_front_len, F_front_sum, F_middl_lst, F_middl_len, F_middl_sum, F_back__lst, F_back__len, F_back__sum, F_intron]
pd.DataFrame(ex_ref_DF)
```

이제 이걸 기존 DataFrame에 적용시켜야합니다  
이를 위해선 lambda를 써야하고 이걸 map으로 적용시켜서 DataFrame로 만들면 됩니다  
변수는 총 6개로 refFlat_DF['strand'], refFlat_DF['exnSta'], refFlat_DF['exnEnd'], refFlat_DF['cdsSta'], refFlat_DF['cdsEnd']  
받을 시 쓸 colums은 ['exnList', 'exnLenList', 'exnLen', 'utr5','utr5LenList','utr5Len','cdsList','cdsLenList','cdsLen','utr3','utr3LenList','utr3Len','intron']로 만듭니다

```python
sep_UTR = pd.DataFrame(map(lambda x1,x2,x3,x4,x5,x6:sep_exn(x1,x2,x3,x4,x5,x6),\
                               refFlat_DF['strand'], refFlat_DF['exnSta'], refFlat_DF['exnEnd'], refFlat_DF['exnList'], refFlat_DF['cdsSta'], refFlat_DF['cdsEnd']),\
                               columns=['utr5','utr5LenList','utr5Len','cdsList','cdsLenList','cdsLen','utr3','utr3LenList','utr3Len','intron'])
sep_UTR
```

sep_UTR을 refFlat_DF옆에다 붙일 것이므로

```python
refFlat_DF = refFlat_DF.merge(sep_UTR,right_index=True,left_index=True)
refFlat_DF
```
